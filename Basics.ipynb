Inputs:
- Hours Studied
- Attendance

Output:
- Pass or Fail

A Neural Network learns how these inputs influence the output.

Instead of manually writing rules,
the network learns those rules from data.


## Main Components:

1. Inputs
2. Neurons
3. Weights
4. Bias
5. Activation Function
6. Hidden Layers
7. Output Layer


## 1.What is an Input?

Inputs are the information given to the network.

Example:

Student Prediction Model

Input 1 = Hours Studied
Input 2 = Attendance Percentage

## 2.What is a Neuron?

A neuron is the basic building block of a neural network.

Think of it as a tiny decision-maker.

- Neuron asks: "Based on study hours and attendance, do I think this student is likely to pass?"



## 3. What are Weights?

Weights tell the network:

"How important is each input?"

- Hours Studied   Weight = 0.9
- Attendance      Weight = 0.2

This means:

Study hours are more important than attendance.

##### Analogy:

When deciding whether someone will pass an exam:

Study Hours may matter a lot
Attendance may matter a little

Weights represent that importance.

## 4. What is Bias?

Bias is an extra adjustable value.

Think of it as a starting opinion.

- Teacher's opinion:

"I generally think students pass unless evidence says otherwise."

Bias allows the network to shift predictions up or down.

## 5. What is an Activation Function?

After a neuron combines all information, it decides:

- Should I activate strongly?
- Should I activate weakly?
  
Activation functions help make that decision.

- Without activation functions: The network becomes too simple.

- With activation functions: The network can learn complex patterns.

Common activations:

- ReLU
- Sigmoid
- Tanh

## 6. What is a Hidden Layer?

Hidden layers sit between inputs and outputs.

Input Layer --> Hidden Layer --> Output Layer

Why "hidden"?

Because we cannot directly see what each neuron is learning.

##### Example:

Input:
- Hours Studied
- Attendance

Hidden neurons may automatically learn:

- Study Discipline
- Exam Readiness
- Consistency

Nobody explicitly tells them to learn these concepts.

They discover useful patterns during training.

## What is the Output Layer?

The output layer produces the final prediction.

Examples:

- Spam Detection

0 = Not Spam
1 = Spam

- Student Prediction

0 = Fail
1 = Pass

- Loan Model

Probability of Default = 0.15

```python
mermaid_code1="""
flowchart LR

%% Input Layer
X1[Feature 1]
X2[Feature 2]

%% Hidden Layer 1
H11((H1))
H12((H2))

%% Hidden Layer 2
H21((H1))
H22((H2))

%% Output Layer
O((Output))

%% Input -> Hidden Layer 1
X1 -->|w1| H11
X1 -->|w2| H12

X2 -->|w3| H11
X2 -->|w4| H12

%% Hidden Layer 1 -> Hidden Layer 2
H11 -->|w5| H21
H11 -->|w6| H22

H12 -->|w7| H21
H12 -->|w8| H22

%% Hidden Layer 2 -> Output
H21 -->|w9| O
H22 -->|w10| O

%% Binary Classification
O --> D{Probability > 0.5?}

D -->|Yes| C1[Class 1]

D -->|No| C0[Class 0]
"""

render_mermaid(mermaid_code1) 
```
